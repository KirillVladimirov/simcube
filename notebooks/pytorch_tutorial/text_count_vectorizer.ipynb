```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns
%matplotlib inline
```

```python
rcParams['figure.figsize'] = 16, 10
```

```python
corpus = [
    'Time flies flies like an arrow.',
     'Fruit flies like a banana.'
]
```

```python
one_hot_vectorizer = CountVectorizer(lowercase=True, binary=True)
```

```python
one_hot = one_hot_vectorizer.fit_transform(corpus)
one_hot
```

```python
one_hot.toarray()
```

```python
one_hot_vectorizer.get_feature_names()
```

```python
sns.heatmap(one_hot.toarray(), annot=True, cbar=True, xticklabels=one_hot_vectorizer.get_feature_names(), yticklabels=['Предложение 1', 'Предложение 2'])
```
