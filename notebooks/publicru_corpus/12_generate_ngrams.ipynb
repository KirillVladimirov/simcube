```python
from nltk.corpus import webtext
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
```

```python
import nltk
nltk.download('webtext')
```

```python
webtext.words("pirates.txt")[:100]
```

```python
text_words = [w.lower() for w in webtext.words("pirates.txt")]
```

```python
text_words
```

```python
len(text_words)
```

```python
finder = BigramCollocationFinder.from_words(text_words)
```

```python
finder.nbest(BigramAssocMeasures.likelihood_ratio, 10)
```

```python
from pymongo import MongoClient
import pandas as pd
import matplotlib.pyplot as plt
```

```python
client = MongoClient('localhost', 27017)
db = client.publicru_test
collection = db.documents_collection
```

```python
result = collection.find({}).skip(1000).limit(1)
document = result[0]
print(document["body"], "\n\n")
```

```python
document["t_body"]
```

```python
list(finder(['more', 'is', 'said', 'than', 'done']))
```

```python
list(nltk.bigrams(['more', 'is', 'said', 'than', 'done']))
```

```python
text = []
bigrams = []
trigrams = []
for para in document["t_body"][0]:
    for sent in para:
        bigrams += list(nltk.bigrams(sent))
        trigrams += list(nltk.trigrams(sent))
        text += sent
```
