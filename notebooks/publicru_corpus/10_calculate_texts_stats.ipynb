```python
def textstat(t_text):
    vocab = nltk.FreqDist()
    for para in t_text[0]:
        for sent in para:
            for word in sent:
                vocab[word] += 1
    paras_count = len(t_text[0])
    sents_count = sum([len(para) for para in t_text[0]])
    words_count = sum([len(sent) for para in t_text[0] for sent in para])
    hapaxes = vocab.hapaxes()
    return {
        "paras_count": paras_count,
        "sents_count": sents_count,
        "words_count": words_count,
        "vocab_count": len(vocab),
        "hapax_count": len(hapaxes),
        "hapax_index": round(len(hapaxes) / len(vocab), ndigits=3),
        "ttr": round(len(vocab) / float(words_count), ndigits=3),
        "sppar": round(float(sents_count) / float(paras_count), ndigits=3),
        "wpsen": round(float(words_count) / float(sents_count), ndigits=3),
    }
```
